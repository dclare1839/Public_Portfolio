```python

df = pd.concat([fertility[["Country Name","Country Code"]]]*24, ignore_index=True)
df
```

```python
period = np.arange(2000,2024)
years = np.tile(period, 266)

years = pd.DataFrame(sorted(years), columns=["Year"])
years
```

```python
df = pd.concat([df, years], axis=1)
df
```

```python
# Make a classification to align the annual data into one column
def align_raw_data(raw_df):
    return_df = pd.DataFrame()

    for idx in np.arange(2000,2024):
        return_df = pd.concat([return_df,raw_df[str(idx)]],axis=0, ignore_index=True)
  
    return return_df
```

```python
# Align the data from 2000 to 2023 into one column
df = pd.concat([df,align_raw_data(fertility)], axis=1)
```

```python
df.columns = ['Country Name', 'Country Code', 'Year', 'Fertility']
df
```
